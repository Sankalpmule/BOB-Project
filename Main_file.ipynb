```python
data = pd.read_csv("full_fault_data.csv")
data.drop("Unnamed: 0",axis =1,inplace=True)
data['Started at'] = pd.to_datetime(data['Started at'])
data['Ended at'] = pd.to_datetime(data['Ended at'])
data
```

```python
p1 = data["Fault"].unique().tolist()
errors = {p1[i] : "e{}".format(i+1) for i in range(len(p1))}
atm = data["Terminal ID"].unique().tolist()
count_positive = {e : 0 for e in errors.values()} 
count_total = {e : 0 for e in errors.values()} 
```

```python
dataset = pd.DataFrame() 
# p1 = Fault_data["Fault"].unique().tolist()
# errors = {p1[i] : "e{}".format(i+1) for i in range(len(p1))}
for name in atm:
    print(atm.index(name))
    sub_data = data[data["Terminal ID"] == name][['Started at', 'Ended at', 'Fault']]
    month_start = sub_data.iloc[0,0].date() #dt.datetime(2020, 8, 1, 0, 0, 0)
    month_start = dt.datetime(month_start.year, month_start.month, month_start.day, 0, 0, 0)
    month_end = dt.datetime(2020, 12, 31, 23, 59, 59)
    x = 2
    m = 10
    y = 10
    shift_days = 3
    mu_len = dt.timedelta(days = m, hours = 0, minutes = 0, seconds = 0)  # length of each measurement unit
    ow_gap = dt.timedelta(days = shift_days, hours = 0, minutes = 0, seconds = 0)  # gap between two successive observation windows
    ow_start = month_start
    pred_point = ow_start + dt.timedelta(days = x*m, hours = 0, minutes = 0, seconds = 0)
    pred_end = pred_point + dt.timedelta(days = y, hours = 0, minutes = 0, seconds = 0)

    error_count = {e : 0 for e in errors.values()}
    rows = []
    target = []

    while(pred_end <= month_end):
        mu_start = ow_start
        mu_end = mu_start + mu_len
        single_row = []
        colnames = []
        
        for mu_no in range(x):
        
            s = sub_data.loc[((sub_data["Started at"] >= mu_start) & (sub_data["Started at"] < mu_end)) | ((sub_data["Started at"] < mu_start) & (sub_data["Ended at"] > mu_start))]
            
            
            
            error_mu = list(s["Fault"])
            for e in error_mu:
                error_count[e] += 1 
            mu_start = mu_end
            mu_end = mu_end + mu_len
            single_row.extend(list(error_count.values()))
            error_count = {e : 0 for e in errors.values()} 
            for e in errors.values():
                colnames.append(e + ".m{}".format(mu_no + 1))
        
        s = sub_data.loc[((sub_data["Started at"] >= ow_start) & (sub_data["Started at"] < pred_point)) | ((sub_data["Started at"] < ow_start) & (sub_data["Ended at"] > ow_start))]
       # s1 = sub_data[sub_data[:,start_index] >= ow_start]
       # s1 = s1[s1[:,start_index] < pred_point]

       # s2 = sub_data[sub_data[:,start_index] < ow_start]
       # s2 = s2[s2[:,end_index] > ow_start]
        
       # s = np.concatenate([s1,s2])
        
        t = ticket.loc[(ticket["ATM ID"] == name) & (ticket["Ticket start"] <= pred_end) & (ticket["Ticket start"] >= mu_start)]
      #  t = t[t["Ticket start"] <= pred_end]
      #  t = t[t["Ticket start"] >= mu_start]
            
        for e in errors.values():
            v = [] 
            dist_from_pred = pd.Series(pred_point - (s[s["Fault"] == e]["Started at"])).apply(lambda x : (x.seconds)/60)
            if dist_from_pred.shape[0] != 0:
                single_row.extend([round(dist_from_pred.min(),2),round(dist_from_pred.max(),2),round(dist_from_pred.mean(),2)])
            #    dist_from_pred = list(dist_from_pred)
                dist_from_pred = list(dist_from_pred.sort_values(ascending = False))
                for i in range(len(dist_from_pred)-1):
                    v.append(dist_from_pred[i] - dist_from_pred[(i+1)])
                if len(v) == 0:
                    single_row.extend([0,0])
                else:
                    single_row.extend([round(np.mean(v),2), round(np.std(v),2)])
            else:
                single_row.extend([0,0,0,0,0])   
                
            colnames.extend(["min(D{})".format(e[1:]),"max(D{})".format(e[1:]),"mean(D{})".format(e[1:]),"mean(V{})".format(e[1:]),"std(V{})".format(e[1:])])
    
            count = s[s["Fault"] == e].shape[0]
            count_total[e] = count_total[e] + count
           # count_total[name][e] = count_total[name][e] + count
            if t.shape[0] > 0:
                count_positive[e] = count_positive[e] + count
              #  count_positive[name][e] = count_positive[name][e] + count
        
        target.append(0 if t.shape[0] == 0 else 1)

        rows.append(single_row)
        
        ow_start = ow_start + ow_gap
        pred_point = ow_start + dt.timedelta(days = x*m, hours = 0, minutes = 0, seconds = 0)
        pred_end = pred_point + dt.timedelta(days = y, hours = 0, minutes = 0, seconds = 0)    
        
    profile = pd.DataFrame([name for i in range(len(rows))], columns = ["ATM ID"])
    df = pd.DataFrame(rows, columns = colnames)
    target_df = pd.DataFrame(target, columns = ["Target"])
    df = pd.concat([profile, df, target_df], axis = 1)
    
    dataset = pd.concat([dataset, df])   
```

```python
print(classification_report(y_test, clf.predict(X_test)))
```

```python
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression(random_state=0)
lg.fit(X_train, y_train)
y_pred = lg.predict(X_test)
print(classification_report(y_test, y_pred))
```
